```python
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df_original=dataset['train'].to_pandas()
df_altered=df_original.copy()
df_original.index.name='job index'
df_original

df_DA_Us = df_altered[(df_altered['job_country'] == 'United States') & (df_altered['job_title_short'] == 'Data Analyst')]
```

```python
df_plot = df_DA_Us['job_location'].value_counts().head(10).to_frame()
sns.set_theme(style="ticks")
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('number of jobs per job title')
plt.xlabel('Number of jobs')
plt.ylabel('')
plt.show()
```

```python
fig, ax = plt.subplots(1,3)
fig.set_size_inches((12,5))
dict={
    'job_work_from_home': 'Work from home',
    'job_no_degree_mention': 'Job Degree Ref',
    'job_health_insurance': 'Health insurance offered'
}
for i, (column, title) in enumerate(dict.items()):
    ax[i].pie(df_DA_Us[column].value_counts(), startangle=90, autopct='%1.1f%%', labels= [False, True])
    ax[i].set_title(title)
plt.show()
```

```python
df_plot = df_DA_Us['company_name'].value_counts().head(10).to_frame()
sns.set_theme(style="ticks")
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('number of companies per job title')
plt.xlabel('Number of jobs')
plt.ylabel('')
plt.show()
```
